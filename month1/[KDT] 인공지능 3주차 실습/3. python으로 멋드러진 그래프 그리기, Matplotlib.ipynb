## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- `%matplotlib inline`을 통해서 matplotlib을 사용하여 시각화된 결과를 셀 안에서 보여달라는 명령이다.

cf) 라이브러리 vs 프레임워크

라이브러리는 개발자들이 만들었을 뿐 우리가 이를 바탕으로 원하는 목표를 달성하기 위해서는 우리가 라이브러리 내부에 있는 코드들을 조합해서 결과를 내야한다. ex) numpy, pandas, matplotlib  
프레임워크의 경우에는 이미 틀이 짜여있고 우리는 그 틀에서 내용물을 채면서 결과물을 완성하는 구조로 되어있다. ex) flask, django

```python
import numpy as np
import pandas as pd
# matplotlib에서 제일 많이 사용하는게 파이썬으로 plotting을 해줄 수있는 
# pyplot모듈을 가장 많이 사용한다. 애에 대한 alias는 plt이다
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study

```python
# 기본적인 plot함수는 꺾은선 그래프를 그리는 함수라는 사실을 알 수있다.
# 실제 plotting을 하는 함수 y = x + 1
plt.plot([1,2,3,4,5])

# plt를 확인하는 명령
plt.show()

# plt.show()를 사용하지 않아도 현재 인터프리터 환경이기 때문에 그래프를 나오게 할 수있지만 그럼
# 에도 plt.show()를 적어줌으로서 그래프를 명시적으로 보겠다고 써주면
# 그래프 객체의 이름이 나오지 않고 그래프가 보여지게 된다.
```

### Figsize: Figure(도면)의 크기 조절

```python
# 그래프의 도면 크기를 figsize인자를 통해 설정가능, 값은 튜플 형태로 넣는다.
plt.figure(figsize = (3,3))
plt.plot([1,2,3,4,5])
plt.show()
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y = x를 그려보면:

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
x = np.array([1,2,3,4,5])# 정의역
y = np.array([1,4,9,16,25]) # f(x)

plt.plot(x,y)
plt.show()
```

```python
# np.arrange()를 사용하여 그래프를 좀 더 smooth하게 나타내 보자!
# numpy라이브러리를 사용하여 만든 데이터들도 matplotlib을 통해서 나타낼 수있다!

x = np.arange(-10,10,0.1)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
# x, y축에 설명추가하기

x = np.arange(-10,10,0.1)
y = x**2

plt.xlabel("x value")
plt.ylabel("y value")
plt.title("y = x**2")

plt.plot(x,y)
plt.show()
```

```python
# x, y축의 범위를 설정하기

x = np.arange(-10,10,0.1)
y = x**2

plt.xlabel("x value")
plt.ylabel("y value")
plt.title("y = x**2")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
# 이렇게 lim을 사용해서도 가능!
# plt.xlim(-5,5)

plt.plot(x,y)
plt.show()
```

```python
# x, y축에 눈금 설정하기

x = np.arange(-10,10,0.1)
y = x**2

plt.xlabel("x value")
plt.ylabel("y value")

# 그래프에 제목 넣기
plt.title("y = x**2")

plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
# 이렇게 lim을 사용해서도 가능!
# plt.xlim(-5,5)

plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정
plt.yticks([i for i in range(0, 28, 3)]) # y축의 눈금 설정
# 반드시 plt.legend를 그리는 시점은 그래프를 그린 이후가 되어야한다!
# 범례(legend)를 표시하려면 먼저 plt함수 인자에 label를 넣어준다음에 plt.legend()를 실행해야한다!
plt.plot(x,y, label = "trend")
plt.legend()

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프(Plot)

```python
# import numpy
# numpy.random.randint(a, b, ...)는 [a, b) 의 범위 중 하나의 수를 임의로 추출합니다.

# import random
# random.randint(a, b)는 [a, b]의 범위 중 하나의 수를 임의로 추출합니다.

x = np.arange(20)
y = np.random.randint(0, 21, 20) # 난수를 0부터 20까지 20번 생성

plt.axis([ 0, 20, 0,20 ])
plt.yticks([i for i in range(0, 21, 5)])
plt.plot(x, y)
plt.show()

# y축을 20까지 보이게 하고싶다면?, y축을 "5"단위로 보이게하고 싶다면?
```

### 산점도 (Scatter Plot)

- `.scatter()`

```python
# 산점도는 두 변수사이의 상관관계를 볼 때 주로 사용한다.
plt.scatter(x,y)
plt.show()
```

### 박스 그림(Box Plot)

- 수치형 데이터에 대한 정보(Q1, Q2, Q3, min, max) 
- `.boxplot()`

```python
plt.boxplot(y)

# 박스 그림의 제일 위에 가로선이 max값, 제일 아래선이 min값,
# min값 선의 그 다음 가로선이 q1(25%)값, 주황선이 중앙값(median)
# 그 다음선이 q3값(75%)이다.
plt.title("box plot of y")
plt.show()

# mission: plot의 title을 box plot of y라고 적자!
```

```python
# x와 y에 대한 boxplot을 동시에 그릴수도 있다.
# 동시에 그릴때는 ()를 추가해서 안에다가 넣어야한다!

plt.boxplot((x,y))
plt.title("box plot of y")
plt.show()
```

### 막대 그래프(Bar plot)
- 범주형 데이터에 많이 사용한다. 범주형 데이터의 값과 그 값의 크기(빈도)를 직사각형으로 나타낸 그림

```python
plt.bar(x,y)

plt.xticks([i for i in range(0,21,1)])
plt.yticks([i for i in range(0,21,2)])
plt.show()
```

```python
# 막대그래프와 유사한 것이 히스토그램!
# 도수분포를 직사각형의 막대 형태로 나타냈다!
# "계급"으로 나타낸 것이 특징: 0,1,2 가아니라 0~2까지의 범주형 데이터로 구성후 그림을 그림


plt.hist(y, bins=np.arange(0,22,2))
plt.xticks([i for i in range(0,21,2)])
# 살펴보면 0에서 2사이의 값이 y개 있고 2~4사이의 값이 y개 있고 이렇게 계급별로 알려준다.
plt.show()
```

### 원형 그래프(Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이하다
- `.pie()`

```python
z = [100,200,300,400]

plt.pie(z, labels = ['one', 'two', 'three', 'four'])
plt.show()
```

### 커널밀도그림(Kernel Density Plot)

- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# in Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0,21,20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot

# 앞에 히스토그램에 있던 분포가 smooth딩하게 결과가 나온것을 확인할 수 있다.
# 히스토그램에 있는 y_value는 도수이고 kdeplot에서는 전체를 1이라고 봤을때 
# 각 부분이 어느정도 density를 가지는지 표현했다고 보면 된다.
# shade라는 인자가 존재한다 이걸 설정하면 그래프의 영역에 음영을 넣어줄 수있다.

sns.kdeplot(y, shade = True)
# sns도 이 명령을 공유한다! 그래서 plt.show()도 sns을 쓸때 같이 써줘야한다!
plt.show()
```

### 카운트그림(Count Plot)

- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name":['Andy','Bob','Cat'],"vote":[True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby(by= vote_df['vote']).count()

vote_count
```

```python
plt.bar(x = [False, True], height=vote_count['name'])
plt.show()
```

```python
# sns의 countplot
# 훨씬더 깔끔하다 라벨 지정도 잘 된 모습이다!!

sns.countplot(x = vote_df['vote'])

plt.show()
```

### 캣 그림(Cat Plot)

- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv('./archive (1)/country_wise_latest.csv')
covid.head(5)
```

```python
# 각 region별로 얼마나 많은 confirmed가 생겼는지 알 수있다.
# 범주형 데이터와 숫자형 데이터를 매핑하는대 탁월하다
# hue라는 인자는 기존에 2차원으로 데이터를 나타냈다면 이제는 각각 점에 범주별로 점들의
# 색깔을 매겨서 실제로는 3개의 혹은 수치형 데이터 간의 관계를 표현할 수 있다.
# catplot은 여러가지 데이터를 한대 묶어서 표현 할 수있다!
# 이게 스트립 그림이다!
s = sns.catplot(x = 'WHO Region',y = 'Confirmed', data = covid, kind = 'strip')
# 아까 matplolib에서 figsize를 조절하는 것과 같은 역할이다!
s.fig.set_size_inches(10,6)
plt.show()
```

```python
# 아까는 strip형태 였고 이제는 violin형태로도 표현 할 수있다!
s = sns.catplot(x = 'WHO Region',y = 'Confirmed', data = covid, kind = 'violin')
# 아까 matplolib에서 figsize를 조절하는 것과 같은 역할이다!
s.fig.set_size_inches(10,6)
plt.show()
```

### 스트립 그림(Strip Plot)

- scatter plot과 유사하기 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
s = sns.stripplot(x = 'WHO Region',y = 'Recovered', data = covid)

plt.show()
```

```python
# cf) swarmplot

# stripplot과 유사한대 위의 그래프처럼 동일한 값을 가지는 값이 많이 존재하는 경우 정확히 
# 점들이 겹치는 경우 점들을 양 옆으로 분산시켜준다 -> 좀더 한눈에 보기 쉽다!

s = sns.swarmplot(x = 'WHO Region',y = 'Recovered', data = covid)

plt.show()
```

### 히트맵(Heatmap)
