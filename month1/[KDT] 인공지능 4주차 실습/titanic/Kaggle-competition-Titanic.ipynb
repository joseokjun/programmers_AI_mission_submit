각 컬럼별 value_counts()를 본 후에 비율이 생존 사망 비율과 크게 차이가 났던 age, class, sex, fare만을 고려한다.
숫자가 높을 수록 산다고 가정하고 비율만큼의 숫자를 모든 승객마다 계산한후 생존 확률값을 더한다.
그 값을 내림차순 정렬하여 값이 큰 38%값을 산다고 나머지는 죽는다고 라벨링(38%는 test데이터에서 38%가 생존했으므로 38%분위수로 나열한것!)

```python
test = pd.read_csv("test.csv")
test
```

```python
test['Survived'] = 0
test
```

```python
test.loc[0,'Fare']
```

```python
def Age_calc(i):
    if (0 <= test.loc[i,'Age']) & (test.loc[i,'Age'] <= 20):
        return 0.46
    elif (20 <= test.loc[i,'Age']) & (test.loc[i,'Age'] <= 40):
        return 0.39
    elif (40 <= test.loc[i,'Age']) & (test.loc[i,'Age'] <= 120):
        return 0.38
    else:
        return 0.38
```

```python
def Pclass_calc(i):
    if test.loc[i,'Pclass'] == 1:
        return 0.39
    elif test.loc[i,'Pclass'] == 2:
        return 0.55
    elif test.loc[i,'Pclass'] == 3:
        return 0.25
```

```python
def Sex_calc(i):
    if test.loc[i,'Sex'] == 'male':
        return 0.19
    elif test.loc[i,'Sex'] == 'female':
        return 0.74
```

```python
def Fare_calc(i):
    if (0 <= test.loc[i,'Fare']) & (test.loc[i,'Fare'] <= 10):
        return 0.2
    elif (10 <= test.loc[i,'Fare']) & (test.loc[i,'Fare'] <= 30):
        return 0.44
    else :
        return 0.5
```

```python
test = pd.read_csv("test.csv")
```

```python
for i in range (len(test)):
    
    test.loc[i,'Survived'] = (0.25*Age_calc(i)) + (0.25*Pclass_calc(i)) + (0.25*Sex_calc(i)) + (0.25*Fare_calc(i))
        
```
