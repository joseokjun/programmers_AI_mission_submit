```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Marvel Superheros/charcters_stats.csv')
df
```

```python
df.sort_values(by = 'Total',ascending = False)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))
sns.kdeplot(x = df[df['Alignment'] == 'good']['Total'], ax=ax)
sns.kdeplot(x = df[df['Alignment'] == 'bad']['Total'], ax=ax)

plt.legend(['good', 'bad'])
plt.show()
```
